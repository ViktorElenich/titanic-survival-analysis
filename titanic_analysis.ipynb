```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.shape        # (строки, столбцы) — размер таблицы
df.info()       # типы данных + где есть пропуски
df.describe()   # статистика по числовым столбцам
```

```python
df['Survived'].value_counts()
```

```python
# только выжившие
survivors = df[df['Survived'] == 1]
survivors.head()

# только пассажиры старше 30
older = df[df['Age'] > 30]
print(older.shape)
```

```python
df.groupby('Sex')['Survived'].mean()
```

```python
df.groupby(['Sex', 'Pclass'])['Survived'].mean()
```

```python
df.pivot_table(values='Survived', index='Sex', columns='Pclass', aggfunc='mean')
```
